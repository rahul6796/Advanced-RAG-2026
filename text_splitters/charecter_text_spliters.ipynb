```python
from termcolor import COLORS, colored
from random import choice
```

```python
colors_list = list(COLORS.keys())[2:8]

colors_list
```

```python
# display the chunks

def display_chunks(chunks):
    colors_list = list(COLORS.keys())[2:8]
    print(f"Total Number of Chunks: {len(chunks)}")
    for num, chunk in enumerate(chunks, 1):
        print(f"Chunk {num}: Length {len(chunk)} chars")
        print(colored(text=chunk, color=choice(colors_list)), end="\n\n")
        
```

```python
display_chunks(chunks)
```

```python
def create_chunks(text: str, chunk_size: int, 
                  separator: str, chunk_overlap: int = 0) -> list[str]:
    # create splitter
    splitter = CharacterTextSplitter(
        chunk_size=chunk_size,
        chunk_overlap=chunk_overlap,
        separator=separator
    )
    # split text
    chunks = splitter.split_text(text=text)
    return chunks
```

```python
display_chunks(create_chunks(text, 100, ""))
```

```python
# increase the chunk size

display_chunks(create_chunks(text, 300, ""))
```

```python
# lets use word boundaries

display_chunks(create_chunks(text, 100, " "))
```

```python
# chunking based on paragraph separator

display_chunks(create_chunks(text, 100, "\n\n"))
```

```python
display_chunks(create_chunks(text, 250, "\n\n"))
```

### Chunk Overlap

```python
def display_chunks_with_overlaps(chunks, chunk_overlap):
    if not chunk_overlap:
        raise(ValueError("Chunk Overlap cannot be 0"))
    
    # create vars
    overlap_text = ""
    overlap_color = "white"
    colors_list = list(COLORS.keys())[2:8]
    
    print(f"Total Number of Chunks: {len(chunks)}")
    for num, chunk in enumerate(chunks, 1):
        print(f"Chunk {num}: Length {len(chunk)} chars")
        if num == 1:
            # print the chunk without overlap
            non_overlap_text = chunk[0:-chunk_overlap]
            overlap_text = chunk[-chunk_overlap:]
            print(colored(non_overlap_text, choice(colors_list)),
                  colored(overlap_text,overlap_color),
                  sep="",end="\n\n")
        else:
            # print the overlap text at start of chunk
            print(colored(overlap_text, overlap_color), end="")
            # print non overlap text
            non_overlap_text = chunk[chunk_overlap:-chunk_overlap]
            print(colored(non_overlap_text, choice(colors_list)), end="")
            # print overlap text at end of chunk
            overlap_text = chunk[-chunk_overlap:]
            print(colored(overlap_text, overlap_color), end="\n\n")
```

```python
display_chunks(chunks=create_chunks(text=text,
                                    chunk_size=100,
                                    separator=""))
```

```python
display_chunks_with_overlaps(chunks=create_chunks(text=text,
                                                  chunk_size=100,
                                                  separator="",
                                                  chunk_overlap=20),
                             chunk_overlap=20)
```

### Token based splitting

```python
# create token based splitter

token_splitter = CharacterTextSplitter.from_tiktoken_encoder(
    encoding_name="cl100k_base",
    chunk_size=50,
    chunk_overlap=5
)
# split the text

token_splitter.split_text(text)
```

### Splitting Documents

```python
from langchain_core.documents import Document
```

```python
# create documents

docs = [Document(
    page_content=text,
    metadata = {"source":"Text on AI"}
)]

print(docs)
```

```python
# create splitter

splitter = CharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=20,
    separator=""
)
```

```python
# split the documents

chunks = splitter.split_documents(docs)

print(chunks)
```
