```python
from dotenv import load_dotenv
import ollama
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import PyPDFLoader
from langchain_ollama.embeddings import OllamaEmbeddings
```

```python
query = "What is Openclaw/Moltbot and what are the major security concerns regarding this tool"
```

```python
query_embeddings = ollama.embed(
    model="embeddinggemma",
    input=query
)
```

```python
query_embeddings["embeddings"]
```

```python
# dimensions

print(len(query_embeddings["embeddings"][0]))
```

```python
query_embeddings = ollama.embed(
    model="embeddinggemma",
    input=query,
    dimensions=512
)
```

```python
# dimensions

print(len(query_embeddings["embeddings"][0]))
```

```python
import os
from pathlib import Path
BASE_DIR = Path(os.getcwd())
PARENT_DIR = BASE_DIR.parent
print(PARENT_DIR)
```

```python
file_path = PARENT_DIR/ "embedding_models" / "Openclaw_Research_Report.pdf"

loader = PyPDFLoader(file_path=file_path)

docs = loader.load()

len(docs)
```

```python
# split documents

chunker = RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=50
)

chunks = chunker.split_documents(docs)

len(chunks)
```

```python
# embed documents

text_documents = [doc.page_content for doc in chunks]

print(len(text_documents))
```

```python
# embed documents

document_embeddings = ollama.embed(
    model="embeddinggemma",
    input=text_documents
)
```

```python

document_embeddings = document_embeddings["embeddings"]
```

```python
# number of embedding vectors

print(len(document_embeddings))
```

```python
# output dimensions

print(len(document_embeddings[0]))
```

## LANGCHAIN

```python
# create the embedder

langchain_embedder = OllamaEmbeddings(
    model="embeddinggemma"
)
```

```python
# embed documents

langchain_document_embeddings = langchain_embedder.embed_documents(
    texts=text_documents
)
```
